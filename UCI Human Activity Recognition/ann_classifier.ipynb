```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("test.csv")
```

```python
features=data.iloc[:,:-2]
```

```python
target=data.iloc[:,-1]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
target = le.fit_transform(target)
```

```python
le.classes_
```

```python
target.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    features,
    target,
    test_size=0.2,
    random_state=42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape  )
```

```python
for i, label in enumerate(le.classes_):
    print(i, "->", label)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()

model.add(Dense(128, activation='relu', input_shape=(561,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(6, activation='softmax'))
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
history = model.fit(
    X_train,
    y_train,
    epochs=10,
    validation_data=(X_test, y_test)
)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)

print("Test Loss:", loss)
print("Test Accuracy:", accuracy)
```

```python
pred = model.predict(X_test)

import numpy as np
predicted = np.argmax(pred, axis=1)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_test, predicted))

print(classification_report(y_test, predicted))
```

```python
# Define different hidden layer architectures to test
# Each tuple represents the nodes in (Layer 1, Layer 2, Layer 3...)
architectures = [
    (64,),                 # 1 Layer: 64 nodes
    (128,),                # 1 Layer: 128 nodes
    (64, 32),              # 2 Layers: 64 nodes then 32 nodes
    (128, 64),             # 2 Layers: 128 nodes then 64 nodes
    (128, 64, 32)          # 3 Layers: 128 -> 64 -> 32 nodes
]

best_accuracy = 0.0
best_config = None

for config in architectures:
    print(f"\n--- Testing Architecture Layout: {config} ---")
    
    # 1. Start a fresh sequential model container
    model = Sequential()
    
    # 2. Add the very first hidden layer with the input shape requirement
    model.add(Dense(config[0], activation='relu', input_shape=(561,)))
    
    # 3. Add any remaining hidden layers dynamically using a loop
    for node_count in config[1:]:
        model.add(Dense(node_count, activation='relu'))
        
    # 4. Add the final output layer (6 classes for your 6 encoded actions)
    model.add(Dense(6, activation='softmax'))
    
    # 5. Compile the loop model configuration
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    
    # 6. Train the model (verbose=0 turns off the massive text spam)
    # Using your updated variable name y_train_encoded
    history = model.fit(X_train, y_train, epochs=10, batch_size=32, 
                        validation_data=(X_test, y_test), verbose=0)
    
    # 7. Extract the validation accuracy score from the final epoch
    current_val_acc = history.history['val_accuracy'][-1]
    print(f"Layout {config} -> Validation Accuracy: {current_val_acc * 100:.2f}%")
    
    # 8. Check if this is our highest performer so far
    if current_val_acc > best_accuracy:
        best_accuracy = current_val_acc
        best_config = config

print("\n🏆 Grid Search Optimization Finished!")
print(f"The best overall layer architecture is: {best_config} with a score of {best_accuracy*100:.2f}%")
```
